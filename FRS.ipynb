```python
from tensorflow.keras.preprocessing import image
```

```python
img = image.load_img("C:/Users/HP/Desktop/fashion/download.png",target_size=(224,224))
```

```python
img
```

```python
img_array = image.img_to_array(img)
```

```python
img_array.shape
```

```python
import numpy as np
expand_image_array = np.expand_dims(img_array,axis=0)
expand_image_array.shape
```

```python
import tensorflow
from tensorflow.keras.preprocessing import image
from tensorflow.keras.layers import GlobalMaxPooling2D
from tensorflow.keras.applications.resnet50 import ResNet50,preprocess_input
```

```python
model = ResNet50(weights='imagenet',include_top=False,input_shape=(224,224,3))
model.trainable = False

model = tensorflow.keras.Sequential([
    model,
    GlobalMaxPooling2D()
])
```

```python
preprocess_image = preprocess_input(expand_image_array)
```

```python
pr = model.predict(preprocess_image).flatten()
```

```python
from numpy.linalg import norm
```

```python
pr/norm(pr)
```
